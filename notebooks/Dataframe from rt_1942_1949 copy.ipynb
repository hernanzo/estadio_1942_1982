```python
from pathlib import Path
import os
import glob
import pickle
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
path = '/Users/hernanadasme/Projects/estadio_1940_1980/ratios_dicts_1942_1949/*.pickle'
rt_1942_1949 = {}

for file_name in glob.glob(path):
    with open(file_name, 'rb') as f:
        my_dictionary = pickle.load(f)
        
        var_name = file_name.split('/')[-1].replace('.pickle', '') # extract file name from path
        rt_1942_1949[var_name] = my_dictionary
```

```python
rt_1942_1949
```

```python
df_rt_1942_1949 = pd.DataFrame.from_dict(rt_1942_1949, orient='index')
```

```python
df_rt_1942_1949
```

```python
df_rt_1942_1949 = df_rt_1942_1949.sort_index(key=lambda x: x.str[-4:])
```

```python
df_rt_1942_1949
```

```python
# create a scatter plot with points instead of bars
fig, ax = plt.subplots(figsize=(40, 10))
for column in df_rt_1942_1949.columns:
    ax.scatter(df_rt_1942_1949.index, df_rt_1942_1949[column], label=column, s=400, alpha = 0.5)

sns.set_style('darkgrid')
sns.set_palette('bright')
sns.set_context('talk')

#ax.legend()
#ax.set_xlabel('Document)
#ax.set_ylabel('Frequency')
ax.set_xticks(df_rt_1942_1949.index)
plt.yticks(fontsize=20)
plt.xticks(rotation=45, fontsize=20)
#plt.legend(fontsize='20')
plt.show()
```

```python
df_rt_1942_1949.to_csv('df_rt_1942_1949.csv', index=True)
```
