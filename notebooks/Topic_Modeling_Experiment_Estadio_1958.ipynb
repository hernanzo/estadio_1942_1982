```python
# read all txt files in the directory into a list of strings
txt_list = []
for file in os.listdir(path):
    if file.endswith('.txt'):
        with open(os.path.join(path, file), 'r') as f:
            txt_list.append(f.read())
```

```python
# create a pandas DataFrame with one row per txt file
df_es_1958 = pd.DataFrame({'page_text': txt_list})
```

```python
df_es_1958
```

```python
# define preprocessing function
def preprocess_text(text):
    # convert text to lowercase
    text = text.lower()
    # remove special characters
    text = re.sub(r'[^\w\s]','',text)
    # remove numbers
    text = re.sub(r'\d+', '', text)
    text = ' '.join([word for word in text.split() if len(word) > 1])
    text = re.sub(r'\^[a-zA-Z]\s+', ' ', text)
    # remove _ characters
    text = text.replace('_', '')
    # remove whitespace
    text = re.sub(r'\s+', ' ', text)
    
    return text
```

```python
# apply preprocessing function to text column
df_es_1958['page_text'] = df_es_1958['page_text'].apply(lambda x: preprocess_text(x))
```

**Preprocessing with Count Vectorizer**

```python
# Define a custom list of Spanish stopwords
spanish_stopwords = nltk.corpus.stopwords.words('spanish')
```

```python
spanish_stopwords
```

```python
# define the list of vowels
vowels = ['a', 'e', 'i', 'o', 'u', 'á', 'é', 'í', 'ó', 'ú']
combinations = [''.join(comb) for comb in itertools.product(vowels, repeat=2)]
combinations
```

```python
letters = list('abcdefghijklmnopqrstuvwxyz')

# generate all possible combinations of the letters and join them into strings
combinationL = [''.join(comb) for comb in itertools.product(letters, repeat=2)]
```

```python
combinationL[:10]
```

```python
list_of_junk = ['fú','guayos', 'conun', 'ca', 'an', 'ados', 'ba', 'aeste', 
                'bí', 'bl', 'aotro', 'jo', 'enel', 'as', 'ael', 'aba', 'aese', 'aios', 'delos', 'dela', 'alladode', 'cadauna', 'cadauno', 'aiul', 'sólo']
```

```python
# iterate over the original list, add one to each element, and append the updated element to the new list
for word in combinations:
    if word not in spanish_stopwords:
        spanish_stopwords.append(word)
for word in list_of_junk:
    if word not in spanish_stopwords:
        spanish_stopwords.append(word)
for word in combinationL:
    if word not in spanish_stopwords:
        spanish_stopwords.append(word)
```

**Fit Count Vectorizer**

```python
cv = CountVectorizer(max_df=0.90, min_df=2, stop_words= spanish_stopwords)
```

```python
dtm = cv.fit_transform(df_es_1958['page_text'])
```

**Latent Dirilecht Allocation**

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
LDA = LatentDirichletAllocation(n_components=10,random_state=300)
```

```python
LDA.fit(dtm)
```
